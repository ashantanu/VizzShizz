### This spiral is an approximation of golden spiral using fibonacci spiral



```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from matplotlib.patches import Rectangle, Arc
from matplotlib.collections import PatchCollection
```

### function to generate fibonacci Seq

```python
#Not the most optimal way but enough for smaller sequences
#Try searching for optimised ways to generate Fibonacci sequences
def fibSeq(n):
    a=0
    b=1
    seq=[a,b]
    n=n-2
    for i in range(n):
        c=a+b
        a=b
        b=c
        seq.append(c)
    return seq
```

### Make the plot

```python
fibonacci=fibSeq(9)
fibonacci.reverse()
rectangles=[]
prev_dim=0
originx=0
originy=0
x=y=0
#plt.xkcd()#to make the plot pretty
fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim(0, fibonacci[0] + 15)
ax.set_ylim(0, fibonacci[0] + 15)
print(fibonacci)
for idx,dim in enumerate(fibonacci):
    if(idx>=len(fibonacci)-1):
    #if(idx>=4):

        continue
    #have to figure out coordinates of 4 boxes, rest is repetition
    x=originx
    y=originy
    if(idx%4)==1:
        y=originy+prev_dim-dim
        x=originx=originx+prev_dim

    if(idx%4==3):
        originy=dim
        
    if(idx%4==2):
        x=originx+prev_dim-dim
    prev_dim=dim
    rect=Rectangle((x,y),dim,dim, antialiased=True, fc='none', ec='b', lw=1,alpha=0.4)
    
    #get the arc
    arc_center_x=x
    arc_center_y=y
    if idx%4==0 or idx%4==3:
        arc_center_x=x+dim
    if idx%4==2 or idx%4==3:
        arc_center_y=y+dim
    arc_center=(arc_center_x,arc_center_y)
    angle=90-(idx%4)*90
    arc=Arc(arc_center,2*dim,2*dim, antialiased=True, fc='b', ec='black', lw=4,angle=angle,theta1=0,theta2=90)

    ax.add_patch(rect)
    ax.add_patch(arc)    
#plt.axis('off')
ax.set_aspect(aspect=1)
#plt.savefig('golden_spiral.png',dpi=800)
```
